```python
import re
import nltk
```

```python
nltk.download("stopwords")  
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
corpus=[]
for i in range(0,len(df['Message'])):
  review=re.sub('[^a-zA-z]',"", df['Message'][i])
  review=review.lower()
  review=review.split()
  cleaned = []
  for word in review:
     if word not in stopwords.words('english'):
        cleaned.append(ps.stem(word))
  review = cleaned
  review=' '.join(review)
  corpus.append(review)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tv = TfidfVectorizer(max_features=2500, ngram_range=(1,2)) # Initialize the TF-IDF Vectorizer
```

```python
X_train = tv.fit_transform(X_train_text).toarray() # Fit TF-IDF on training text and transform
X_test = tv.transform(X_test_text).toarray()     # Transform test text using the fitted TF-IDF
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect_model=MultinomialNB()
spam_detect_model.fit(X_train,y_train) # Use the correctly prepared X_train and y_train
```

```python
y_pred1=spam_detect_model.predict(X_test)
print(accuracy_score(y_test,y_pred1))
```
